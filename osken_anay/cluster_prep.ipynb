```python
import pandas as pd
import numpy as np
```

```python
one_08 = pd.read_csv('2001_2008_player_shot_metrics.csv').rename(mapper={'Unnamed: 1': 'stat_cat'}, axis=1).set_index(['season_end', 'player'])
twenty_09 = pd.read_csv('2009_player_shot_metrics.csv').rename(mapper={'Unnamed: 1': 'stat_cat'}, axis=1).set_index(['season_end', 'player'])
ten_14 = pd.read_csv('2010_2014_player_shot_metrics.csv').rename(mapper={'Unnamed: 1': 'stat_cat'}, axis=1).set_index(['season_end', 'player'])
fift_17 = pd.read_csv('2015_2017_player_shot_metrics.csv').rename(mapper={'Unnamed: 1': 'stat_cat'}, axis=1).set_index(['season_end', 'player'])
eiteen_24 = pd.read_csv('2018_2024_player_shot_metrics.csv').rename(mapper={'Unnamed: 1': 'stat_cat'}, axis=1).set_index(['season_end', 'player'])
master = pd.concat([one_08, twenty_09, ten_14, fift_17, eiteen_24])
master
```

```python
shot_metrics = pd.read_csv('2000_2024_player_shot_metrics.csv', index_col=[0, 1])
shot_metrics
```

```python
advanced = pd.read_csv('2000_2024_player_totals_advanced.csv', index_col=[0, 1])
advanced
```

```python
basics = pd.read_csv('2000_2024_player_totals_basic.csv', index_col=[0, 1])
basics.index.names = ['season_end', 'player']
basics.to_csv('2000_2024_player_totals_basic.csv')
basics
```

```python
cluster_data = basics.join([advanced, master], how='inner')
cluster_data.index.names = ['season_end', 'player']
cluster_data.to_csv('seasonal_clustering_data_master.csv')
```
