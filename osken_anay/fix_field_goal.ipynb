```python
import pandas as pd
```

```python
cluster_data= pd.read_csv('seasonal_clustering_data_master.csv', index_col=[0, 1])
cluster_data
```

```python
basics = pd.read_csv("basics.csv").rename(mapper={'name': 'player'}, axis=1).set_index(['season_end', 'player'])
basics
```

```python
wanted = basics["attempted_field_goals"]
wanted
```

```python
cluster_data = cluster_data.join([wanted], how='inner')
cluster_data.columns
```

```python
cluster_data["Above the Break 3_shots_usage"] = cluster_data["Above the Break 3_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data["Backcourt_shots_usage"] = cluster_data["Backcourt_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data["In The Paint (Non-RA)_shots_usage"] = cluster_data["In The Paint (Non-RA)_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data["Left Corner 3_shots_usage"] = cluster_data["Left Corner 3_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data["Mid-Range_shots_usage"] = cluster_data["Mid-Range_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data["Restricted Area_shots_usage"] = cluster_data["Restricted Area_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data["Right Corner 3_shots_usage"] = cluster_data["Right Corner 3_shots_attempted"] / cluster_data["attempted_field_goals"]
cluster_data
```

```python
cluster_data.drop(columns=["Above the Break 3_shots_attempted", "Backcourt_shots_attempted", "In The Paint (Non-RA)_shots_attempted", "Mid-Range_shots_attempted", "Left Corner 3_shots_attempted", "Restricted Area_shots_attempted", "Right Corner 3_shots_attempted", "attempted_field_goals"], inplace=True)
cluster_data
```

```python
cluster_data.drop(columns=["Above the Break 3_shot_made", "Backcourt_shot_made", "In The Paint (Non-RA)_shot_made", "Mid-Range_shot_made", "Left Corner 3_shot_made", "Restricted Area_shot_made", "Right Corner 3_shot_made"], inplace=True)
```

```python
cluster_data.to_csv("seasonal_clustering_data_master.csv")
```
